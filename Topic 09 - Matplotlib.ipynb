# Introduction

*matplotlib* is a package for graphics creation. The syntax is similar to Matlab on purpose.

```python
import matplotlib.pyplot as plt
```

# Basic graphs 

```python
import numpy as np

x = np.linspace(-5, 50)
y = x ** 2 - 45 * x + 5

plt.plot(x, y, 'r-.')

plt.title('Function', fontsize = 20)
plt.xlabel('x [m]', fontsize = 15)
plt.ylabel('y [m]', fontsize = 15)

plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)

plt.show()
plt.close()
```

## sub plots

```python
plt.figure(figsize = (15,3))

plt.subplot(1,3,1)
plt.plot(x, y)

plt.subplot(1,3,2)
plt.scatter(x, y)

plt.subplot(1,3,3)
plt.scatter(x, -y, c = 'r')

plt.show()
plt.close()
```

```python
fig, axes = plt.subplots(nrows = 2, ncols=2, figsize=(8,8))

axes[0, 0].plot(x, y, c='C0')
axes[0, 1].scatter(x, y, c='C1')
axes[1, 0].plot(x, -y, c='C2')
axes[1, 1].scatter(x, -y, c='C3')

plt.show()
plt.close()
```

## Adding text

Note that you can also add latex notation

```python
plt.plot(x, y)

ymin = y.min()
xmin = x[y.argmin()]

plt.text(xmin, ymin, '|')
plt.text(10, 100, r'$\sum_{i=0}^{10} x^2 - 45 \cdot x + 5$')  

plt.show()
plt.close()
```

## Ploting more than one signal

```python
x = np.linspace(0, 4 * np.pi, 200)

y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)

plt.ylim((-2, 2))
plt.xlim((0, 4 * np.pi))
plt.axhline(0, color='k') 

plt.show()
plt.close()
```

## Adding legends

```python
plt.figure(figsize =(10,4))

plt.plot(x, y1, label = 'sin')
plt.plot(x, y2, label = 'cos')
plt.plot(x, y3, label = 'tan')

plt.legend(fontsize = 15)

plt.ylim((-2, 2))
plt.xlim((0, 4 * np.pi))
plt.axhline(0, color='k') 

plt.show()
plt.close()
```

## Saving figures

Figures can be saved with the function `savefig` in different formats:

* `png`
* `pdf`
* `ps`
* `eps`
* `svg`

```python
x = np.linspace(0, 4 * np.pi, 200)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='sin')
plt.plot(x, y2, label='cos')
plt.legend(title='Function', ncol=2, loc='upper right')
plt.ylim((-1, 1))

plt.savefig('sin_cos.png', dpi=150)

plt.show()
plt.close()
```

```python
import matplotlib.pyplot as plt
import numpy as np

t = np.linspace(0,60e-3,100)

va = np.sin(377*t)
vb = np.sin(377*t - 2*np.pi/3)
vc = np.sin(377*t + 2*np.pi/3)

plt.plot(t,va, label = "Va")
plt.plot(t,vb, label = "Vb")
plt.plot(t,vc, label = "Vc")

tvamax = t[va.argmax()]
plt.text(tvamax, va.max(), 'Max of the phase A') 

plt.legend(loc = 1)
plt.show()
```
